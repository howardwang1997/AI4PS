```python
from rdkit import Chem
from rdkit.Chem import Draw

import random
import json

with open('decoded_all.json', 'r') as f:
    decode = json.load(f)
```

```python
show_idx = random.sample(range(len(decode)), 10)

for i in show_idx:
    # print(f'{i}th decoded scaffold:')
    mol = Chem.MolFromSmiles(decode[i])
    img = Draw.MolToImage(mol)
    img.show()
```
